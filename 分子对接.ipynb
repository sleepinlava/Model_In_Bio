```python
# 2. 执行你的修复代码
!pip install pdbfixer
from pdbfixer import PDBFixer
from openmm.app import PDBFile

print("正在修复 PDB 文件...")

# 确保文件名和你上传的一模一样
fixer = PDBFixer(filename='/content/cleaned_protein.pdb')

fixer.findMissingResidues()
fixer.findMissingAtoms()
fixer.addMissingAtoms()
fixer.addMissingHydrogens(7.0)

# 输出修复后的文件
output_filename = 'OspF_fixed.pdb'
with open(output_filename, 'w') as f:
    PDBFile.writeFile(fixer.topology, fixer.positions, f)

print(f"✅ 修复成功！已生成文件: {output_filename}")
```

```python
# 1. 安装可视化库
!pip install -q py3Dmol

import py3Dmol

# 2. 读取并渲染修复后的 PDB 文件
def visualize_protein(pdb_file):
    with open(pdb_file, 'r') as f:
        pdb_data = f.read()

    # 初始化查看器
    view = py3Dmol.view(width=800, height=600)
    view.addModel(pdb_data, 'pdb')

    # 设置渲染样式：卡通模型显示骨架
    view.setStyle({'cartoon': {'color': 'spectrum'}})

    # 额外显示：把刚才修复的关键残基（如 Lys134）用“棒状”标出
    # OspF 的关键催化残基通常包括 Lys134, His104, Arg146
    view.addStyle({'resi': '104,134,146'}, {'stick': {'colorscheme': 'cyanCarbon'}})

    # 自动居中并缩放
    view.zoomTo()
    return view.show()

# 调用函数
visualize_protein('OspF_fixed.pdb')
```

```python
import pandas as pd
import py3Dmol

# 1. 加载 P2Rank 预测的 CSV 结果
csv_path = '/content/p2rank_2.4.2/test_output/predict_OspF_fixed/OspF_fixed.pdb_predictions.csv'
pockets_df = pd.read_csv(csv_path)

# 修复：移除列名中可能存在的空格（P2Rank 的 CSV 经常有这种格式问题）
pockets_df.columns = [c.strip() for c in pockets_df.columns]

# 2. 读取蛋白 PDB 文件
with open('/content/OspF_fixed.pdb', 'r') as f:
    pdb_data = f.read()

# 3. 初始化 3D 视图
view = py3Dmol.view(width=800, height=600)
view.addModel(pdb_data, 'pdb')
view.setStyle({'cartoon': {'color': 'spectrum'}})

# 4. 将预测的口袋（Pockets）添加为球体
# 我们渲染得分最高的 5 个口袋
for i, row in pockets_df.head(5).iterrows():
    view.addSphere({
        'center': {'x': float(row['center_x']), 'y': float(row['center_y']), 'z': float(row['center_z'])},
        'radius': 4.0,
        'color': 'red',
        'opacity': 0.7
    })
    # 添加标签标注口袋编号
    view.addLabel(f"Pocket {i+1} (Score: {row['score']})", {
        'position': {'x': float(row['center_x']), 'y': float(row['center_y']), 'z': float(row['center_z'])},
        'backgroundColor': 'black',
        'fontColor': 'white'
    })

view.zoomTo()
view.show()
```

```python
import pandas as pd

# 根据你上传的文件名，请确保这里名字是对的
prediction_file = '/content/p2rank_2.4.2/test_output/predict_OspF_fixed/OspF_fixed.pdb_predictions.csv'

print("正在解析口袋地形图...")
try:
    # 1. 读取数据
    df = pd.read_csv(prediction_file)

    # 🔑 核心修复操作：一键清洗所有列名的首尾隐藏空格
    df.columns = df.columns.str.strip()

    # 2. 提取得分最高的第一号口袋 (Pocket 1)
    top_pocket = df.iloc[0]

    # 3. 提取核心数据
    center_x = top_pocket['center_x']
    center_y = top_pocket['center_y']
    center_z = top_pocket['center_z']
    residues = str(top_pocket['residue_ids']).strip()

    print(f"✅ 成功锁定一号靶心 (Pocket 1)！")
    print(f"🎯 坐标中心: X={center_x:.3f}, Y={center_y:.3f}, Z={center_z:.3f}")
    print(f"🧬 包含的残基: {residues[:50]}...") # 打印前50个字符防刷屏

    # 4. 关键催化残基校验
    if '134' in residues:
        print("🔥 完美！这个口袋包含了关键的 Lys134 催化位点，这里就是我们的主战场！")
    else:
        print("💡 提示：这个得分最高的口袋可能是一个未知的变构位点。")

except Exception as e:
    print(f"❌ 读取发生错误: {type(e).__name__} - {e}")
```

```python
# 安装 RDKit（如果没有装过的话）
!mamba install -c conda-forge rdkit -y

from rdkit import Chem
from rdkit.Chem import AllChem

print("正在兵工厂制造测试分子...")
# 我们随便选一个带有一些氢键供体/受体的小分子骨架
smiles = "CC1=C(C=C(C=C1)NC(=O)C2=CC=C(C=C2)CN3CCN(CC3)C)NC4=NC=CC(=N4)C5=CC=CN=C5"
mol = Chem.MolFromSmiles(smiles)

# 给分子加氢并生成 3D 构象
mol = Chem.AddHs(mol)
AllChem.EmbedMolecule(mol, randomSeed=42)
AllChem.UFFOptimizeMolecule(mol)

# 保存为 sdf 格式
Chem.MolToMolFile(mol, 'test_drug.sdf')
print("✅ 武器准备完毕：test_drug.sdf 已生成。")
```

```python
import py3Dmol

print("正在构建 3D 虚拟实验室，加载受体与配体...")

# 1. 读取蛋白和对接结果
with open('/content/OspF_fixed.pdb', 'r') as f:
    protein_data = f.read()

with open('best_docked_pose.sdf', 'r') as f:
    ligand_data = f.read() # 这里面包含了 Gnina 跑出来的所有姿势，默认加载第一个（得分最高的）

# 2. 初始化 3D 视图
view = py3Dmol.view(width=800, height=600)

# 3. 渲染蛋白质（半透明的灰色飘带，方便看清内部）
view.addModel(protein_data, 'pdb')
view.setStyle({'model': 0}, {'cartoon': {'color': 'lightgray', 'opacity': 0.6}})

# 4. 高亮显示我们 P2Rank 找到的靶心口袋残基（用青色标出）
# 注意：这里使用的是你刚才提取出来的残基编号
pocket_residues = '109,120,121,122,129,131,186,188'
view.addStyle({'model': 0, 'resi': pocket_residues}, {'stick': {'colorscheme': 'cyanCarbon'}})

# 5. 渲染药物分子（用醒目的橙色粗棒显示）
view.addModel(ligand_data, 'sdf')
view.setStyle({'model': 1}, {'stick': {'colorscheme': 'orangeCarbon', 'radius': 0.2}})

# 6. 镜头聚焦并展示
view.zoomTo({'model': 1}) # 直接把镜头拉近到药物分子身上
view.show()
```
